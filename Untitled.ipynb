```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\Kathe\CitiBike-2022-2\3.3_Include_Trips.csv",
    index_col=0,
    low_memory=False
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'rideable_type' and count occurrences
rideable_type_counts = df['rideable_type'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(rideable_type_counts, labels=rideable_type_counts.index, autopct='%1.1f%%', startangle=90)

# Title
plt.title('Distribution of Rideable Types')

# Show the plot
plt.show()
```

```python
import plotly.express as px

# Group by 'rideable_type' and 'member_casual', then count the occurrences
usage_counts = df.groupby(['rideable_type', 'member_casual']).size().reset_index(name='count')

# Create bar chart
fig = px.bar(
    usage_counts, 
    x='rideable_type', 
    y='count', 
    color='member_casual', 
    title='Bike Usage by Customer Type (Member vs Casual)',
    labels={'rideable_type': 'Type of Bike', 'count': 'Number of Customers'},
    barmode='stack'  # Stacks the bars for Member vs Casual
)

# Show the plot
fig.show()
```

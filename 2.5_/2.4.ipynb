```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataframe (assuming it's already read as df)
# df = pd.read_csv("your_data.csv")

# Convert 'date' to datetime format for better plotting
df['date'] = pd.to_datetime(df['date'])

# Set the seaborn style
sns.set_theme(style="darkgrid")

# Create the figure and first axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the trips_per_day on the first y-axis
sns.lineplot(x=df['date'], y=df['trips_per_day'], ax=ax1, color='b', label='Trips per Day')
ax1.set_ylabel('Number of Trips', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot the avgTemp on the second y-axis
sns.lineplot(x=df['date'], y=df['avgTemp'], ax=ax2, color='r', label='Average Temperature')
ax2.set_ylabel('Average Temperature (°C)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Title and labels
plt.title('Daily CitiBike Trips vs. Average Temperature')

# Show the legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

**7. Using FacetGrid for Analysis**

```python
df['started_at'] = pd.to_datetime(df['started_at'])
```

```python
df['day_of_week'] = df['started_at'].dt.day_name()  # Gets full weekday names
df['hour_of_day'] = df['started_at'].dt.hour  # Extracts the hour
```

```python
heatmap_data = df.pivot_table(index='day_of_week', columns='hour_of_day', values='trips_per_day', aggfunc='sum')
```

```python
heatmap_data = heatmap_data.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
plt.figure(figsize=(14, 7))  # Adjust figure size
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=False)  # Remove annotations

plt.title("Heatmap of Trip Counts by Hour and Day of the Week", fontsize=16)
plt.xlabel("Hour of the Day", fontsize=14)
plt.ylabel("Day of the Week", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('heatmap_image.png')
plt.close()
plt.show()
```

**This heatmap is a useful visualization for analyzing the times of day when Citibikes are most frequently used throughout the week. The heatmap clearly highlights the busiest hours, showing that bike usage peaks during the evening hours, especially in the middle of the week (Tuesday through Thursday). The heatmap uses color gradients, where darker hues represent higher usage, and lighter hues indicate lower usage. For example, the very light yellow hue on the left side of the graph signifies the lowest usage, occurring during the early morning hours across all days.**
